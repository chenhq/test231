```python
def construct_industry_tree(industry2code, stks_ohlcv, weight_column='negMarketValue', default_weight = 1e8, weight_scale = 1e8, verbose=False):
    industry_tree = []
    for index, value in industry2code.iterrows():
        ticker = value['ticker']
        if ticker in stks_market and len(stks_market[ticker]) > 0:
            weight = stks_market[ticker].iloc[-1][weight_column] // weight_scale
        else:
            if verbose:
                print('ticker: {} not found!'.format(ticker))
            continue

        found = False
        for sub1 in industry_tree:
            if sub1['name'] == index[0]:
                found = True
                sub1['value'] += weight
                sub2s = sub1['children']
                break
        if not found:
            sub2s = []
            industry_tree.append({'value': weight, 'name': index[0], 'children': sub2s})

        found = False
        for sub2 in sub2s:
            if sub2['name'] == index[1]:
                found = True
                sub2['value'] += weight
                sub3s = sub2['children']
                break
        if not found:
            sub3s = []
            sub2s.append({'value': weight, 'name': index[1], 'children': sub3s})

        found = False
        for sub3 in sub3s:
            if sub3['name'] == index[2]:
                found = True
                sub3['value'] += weight
                stks = sub3['children']
                break
        if not found:
            stks = []
            sub3s.append({'value': weight, 'name': index[2], 'children': stks})

        found = False
        for stk in stks:
            if stk['name'] == value['secShortName']:
                found = True
                break
        if not found:
            stks.append({'value': weight, 'name': value['secShortName'], 'ticker': ticker})
    return industry_tree
```

```python
industry_tree_by_turnover = construct_industry_tree(industry2code, stks_ohlcv, weight_column='turnoverValue')
```

```python
industry_tree_by_negMarketValue = construct_industry_tree(industry2code, stks_ohlcv, weight_column='negMarketValue')
```

```python
def ohlcv_pct_chg(ohlcv, by='negMarketValue'):
    pre_close = ohlcv['close'].shift(1).bfill()
    result = ohlcv[by]
    new_cols = [by]
    ohlcv_cols = ['open', 'high', 'low', 'close']
    for col in ohlcv_cols:
        new = ((ohlcv[col] / pre_close) - 1) * ohlcv[by]
        result = pd.concat([result, new], axis=1)
        new_cols.append(col)
    result.columns = new_cols
    return result
```

```python
list_ohlcv_by_negMarketValue = []
for stk in stks_ohlcv:
    list_ohlcv_by_negMarketValue.append(ohlcv_pct_chg(stks_ohlcv[stk]))
```

```python
def ohlcv_index_by_negMarketValue(ohlcv):   
    industry_stks = pd.pivot_table(ohlcv, index=['date'], columns=['code']).fillna(0)
    ohlcv_by_negMarketValue = pd.DataFrame()
    negMarketValue_sum = industry_stks.loc[:, Idx['negMarketValue', :]].sum(axis=1)
    negMarketValue_sum.name = 'negMarketValue'
    ohlcv_by_negMarketValue = pd.concat([ohlcv_by_negMarketValue, negMarketValue_sum], axis=1)
    for col in ['open', 'high', 'low', 'close']:
        # print('col:', col)
        mean = industry_stks.loc[:, Idx[col, :]].sum(axis=1) / negMarketValue_sum
        mean.name = col
        # print(mean)
        ohlcv_by_negMarketValue = pd.concat([ohlcv_by_negMarketValue, mean], axis=1)
    return ohlcv_by_negMarketValue
```

```python
insdustry_ohlcv = defaultdict(lambda: defaultdict(lambda: defaultdict(dict)))
industry_levels = ['industryName1', 'industryName2', 'industryName3']
tag = 'ohlcv_by_negMarketValue'
insdustry_ohlcv[tag] = ohlcv_index_by_negMarketValue(ohlcv_by_negMarketValue_industry)
for level in range(len(industry_levels)):
    for keys, group in ohlcv_by_negMarketValue_industry.groupby(industry_levels[:level+1]):
        print(keys)
        if level == 0:
            keys = [keys]
        for i, key in enumerate(keys):
            # print('key:', key)
            if i == 0:
                nested = insdustry_ohlcv[key]
            else:
                nested = nested[key]
            
            if i == len(keys) -1:
                nested[tag] = ohlcv_index_by_negMarketValue(group)
```
